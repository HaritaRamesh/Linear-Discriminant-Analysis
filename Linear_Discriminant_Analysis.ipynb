```
#reshaping the means
m0 = m0.reshape(784)
m1 = m1.reshape(784)
m2 = m2.reshape(784)
m3 = m3.reshape(784)
m4 = m4.reshape(784)
m5 = m5.reshape(784)
m6 = m6.reshape(784)
m7 = m7.reshape(784)
m8 = m8.reshape(784)
m9 = m9.reshape(784)
```

```
#Part 2:
#Creating an array of means : 
means = np.array([m0,m1,m2,m3,m4,m5,m6,m7,m8,m9])
cov_lst = np.array([c0,c1,c2,c3,c4,c5,c6,c7,c8,c9])
```

```
#Calculating prior
#prior = count of data belonging to a particular class / count of all the data
list_of_class = [X0, X1, X2, X3, X4, X5, X6, X7, X8, X9]
priors = []
for each_priors in list_of_class:
  priors.append(len(each_priors)/60000)
```

$W_{i} = −0.5*∑^{-1}_{i}$


 $N_{i} = ∑_{i}^{-1}μ_{i}$

$B_{i0} = -0.5μ_{i}^{t}∑_{i}^{-1}μ_{i} + lnP(w_{i}) - 0.5ln|∑_{i}| $

```

from numpy.linalg import pinv
W_i=[] 
N_i = []
B_i = []
i=0

for ind_c in cov_lst:
  W_i.append(-0.5*pinv(ind_c))
  N_i.append(np.matmul(pinv(ind_c),means[i] ))
  B_i.append((-0.5*np.matmul(np.matmul(means[i].T,pinv(ind_c)),means[i])) + np.log(priors[i]) - (0.5*np.log(np.linalg.det(ind_c)+0.1)))
  i=i+1
```

```
X_test=X_test.reshape(10000,784)
```

$g_{i}(x)= x^tW_ix + N_{i}^{t}x + B_{i0}$

```
# Quadratic discriminant function equation 
# gi(x)=xtWix+Ntix+Bi0

num_of_classes=10
g_i = []
i=0
for data in X_test:
  for each_class in range(num_of_classes):
    g_i.append((np.matmul(np.matmul(data.T,W_i[each_class]),data)) + np.matmul(N_i[each_class].T,data) + B_i[each_class])
```

$λ(α_{i}|ω_{j})={0,i=j;}$
               ${1,i≠j}$

```
#Calculating zero - one loss :
#zero-one loss/symmetric loss function λ(αi|ωj)={0, if i=j or 1, if i≠j,
num_of_classes=10
i=0
error=0
for data in X_test:
  prob_list = []
  for each_class in range(num_of_classes):
    g = (np.matmul(np.matmul(data.T,W_i[each_class]),data)) + np.matmul(N_i[each_class].T,data) + B_i[each_class]
    prob_list.append(g)
  #print(".....................")
  if y_test[i]!=np.argmax(prob_list):
    error=error+1
  i=i+1

```

```
#Calculating accuracy
print("Accuracy: ", (len(X_test)-error)/len(X_test)*100)
```
